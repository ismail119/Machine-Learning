```python
import pandas as pd,seaborn as sns,numpy as np

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
data = pd.read_csv('spam.csv')
```

```python
data = data.iloc[:,:2]
```

```python
data
```

```python
data.isnull().sum()
```

```python
#spam=1 , ham=0
data.v1 = data.v1.apply(lambda x: 0 if x=="ham" else 1)
```

```python
data.head()
```

```python
data.columns=["label","sms"]
```

```python
df = data.copy()
```

```python
#removing punctions
df.sms = df.sms.str.replace("[^\w\s]","")
```

```python
#removing numbers
df.sms = df.sms.str.replace("\d","")
```

```python
#letter make lower
df.sms = df.sms.apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
import nltk
from nltk.corpus import stopwords
from textblob import Word
from textblob import TextBlob
```

```python
stopwords = stopwords.words("english")
stopwords.extend(['u','nah','dont'])
```

```python
#Removing Stopwords
df.sms = df.sms.apply(lambda x: " ".join(x for x in x.split() if x not in stopwords))
```

```python
df.head()
```

```python
#Lemmatization
df.sms = df.sms.apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
```

```python
#Variable splitting
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(df["sms"],df["label"],test_size=0.3,random_state=42)
```

```python
#TF-IDF counting operation
from sklearn.feature_extraction.text import TfidfVectorizer

tf_idf_vectorizer = TfidfVectorizer().fit(x_train)
```

```python
x_train_TfIdf = tf_idf_vectorizer.transform(x_train)
x_test_TfIdf = tf_idf_vectorizer.transform(x_test)
```

```python
x_test_TfIdf.toarray()
```

# Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
```

```python
models = dict(
    lr_model = LogisticRegression(),
    rf_model = RandomForestClassifier(),
    lgbm_model = LGBMClassifier(),
    cb_model = CatBoostClassifier()
)
```

```python
#fitting 
for model in models.values():
    model.fit(x_train_TfIdf,y_train)
```

```python
#scoring
for k,v in models.items():
    print("{} has score=> {}".format(k,v.score(x_test_TfIdf,y_test)))
```
