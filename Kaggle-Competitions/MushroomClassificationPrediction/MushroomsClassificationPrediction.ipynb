```python
import pandas as pd,numpy as np,seaborn as sns
```

```python
data = pd.read_csv('mushrooms.csv')
```

```python
data.info()
```

```python
for i in data.columns:
    if len(data[i].unique())==2:
        print(i,data[i].unique())
```

```python
df = data.copy()
```

```python
#delete just one value variable because it is not variable :)
df.drop("veil-type",axis=1,inplace=True)
```

```python
#label encoding binary values

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df["class"] = encoder.fit_transform(df["class"])
class_mapping = {label: index for index ,label in enumerate(encoder.classes_)} 


df["bruises"] = encoder.fit_transform(df["bruises"])
bruises_mapping = {label: index for index ,label in enumerate(encoder.classes_)} 


df["gill-attachment"] = encoder.fit_transform(df["gill-attachment"])
gill_attachment_mapping = {label: index for index ,label in enumerate(encoder.classes_)} 


df["gill-spacing"] = encoder.fit_transform(df["gill-spacing"])
gill_spacing_mapping = {label: index for index ,label in enumerate(encoder.classes_)} 


df["gill-size"] = encoder.fit_transform(df["gill-size"])
gill_size_mapping = {label: index for index ,label in enumerate(encoder.classes_)} 


df["stalk-shape"] = encoder.fit_transform(df["stalk-shape"])
stalk_shape_mapping = {label: index for index ,label in enumerate(encoder.classes_)} 
```

```python
df.dtypes[df.dtypes=="object"].index
```

```python
#One hot encoding

df_ = pd.get_dummies(data=df,columns=['cap-shape', 'cap-surface', 'cap-color', 'odor', 'gill-color',
       'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring',
       'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-color',
       'ring-number', 'ring-type', 'spore-print-color', 'population',
       'habitat'],prefix=['cap-shape', 'cap-surface', 'cap-color', 'odor', 'gill-color',
       'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring',
       'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-color',
       'ring-number', 'ring-type', 'spore-print-color', 'population',
       'habitat'])
```

```python
X = df_.drop('class',axis=1)
y = df_["class"] 
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = dict(
    lr_model = LogisticRegression(),
    cb_model = CatBoostClassifier(),
    lgbm_model = LGBMClassifier(),
    rf_model = RandomForestClassifier()
)
```

```python
for model in models.values():
    model.fit(x_train,y_train)
```

```python
for k,v in models.items():
    print("{} has score => {}".format(k,v.score(x_test,y_test)))
```

```python
from sklearn.metrics import accuracy_score

for k,v in models.items():
    print("{} has score => {}".format(k,accuracy_score(v.predict(x_test),y_test)))
```
