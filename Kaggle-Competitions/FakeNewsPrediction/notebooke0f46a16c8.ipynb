```python
exceptions = df.date.str.count(',')!=1
```

```python
df.loc[exceptions,'date'] = ",2018"
```

```python
#let see is date has any colleration between fake situation.(year approach)

df["year_of_subject"] = df.date.apply(lambda x: x.split(',')[1])
```

```python
df.year_of_subject.unique()
```

```python
df.year_of_subject = df.year_of_subject.apply(lambda x: x.strip())
```

```python
df.year_of_subject.unique()
```

```python
#fuzzy data
bad_data = df.title.str.startswith("https")
```

```python
df = df[~bad_data]
```

```python
bad_data = df.title.str.startswith("Homepage")
```

```python
df = df[~bad_data]
```

```python
df.drop("date",axis=1,inplace=True)
```

# Now lets NLP

```python
import nltk
from nltk.corpus import stopwords
from textblob import Word
from textblob import TextBlob
```

```python
df['concatinateText'] = df['title'] + ' ' + df['text']
```

```python
#lower letter
df.concatinateText = df.concatinateText.apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
#removing punctions
df.concatinateText = df.concatinateText.str.replace("[^\w\s]","")
```

```python
#removing numerical values
df.concatinateText = df.concatinateText.str.replace("\d","")
```

```python
stopwords = stopwords.words("english")
df.concatinateText = df.concatinateText.apply(lambda x: " ".join(x for x in x.split() if x not in stopwords))
```

```python
df.concatinateText = df.concatinateText.apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
```

```python
df.drop(["text","title"],axis=1,inplace=True)
```
