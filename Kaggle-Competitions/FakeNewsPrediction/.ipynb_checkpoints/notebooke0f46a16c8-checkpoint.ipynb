```python
import nltk
from nltk.corpus import stopwords
from textblob import Word
from textblob import TextBlob
```

```python
#lower letter
df.title = df.title.apply(lambda x: " ".join(x.lower() for x in x.split()))
df.text = df.text.apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
#removing punctions
df.title = df.title.str.replace("[^\w\s]","")
df.text= df.text.str.replace("[^\w\s]","")
```

```python
#removing numerical values
df.text = df.text.str.replace("\d","")
df.title = df.title.str.replace("\d","")
```

```python
stopwords = stopwords.words("english")
df.title = df.title.apply(lambda x: " ".join(x for x in x.split() if x not in stopwords))
df.text = df.text.apply(lambda x: " ".join(x for x in x.split() if x not in stopwords))
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
def lemmatize_word(word):
    return nlp(word)[0].lemma_
```

```python
df.text = df.text.apply(lambda x: ' '.join([lemmatize_word(word) for word in x.split()]))
df.title = df.title.apply(lambda x: ' '.join([lemmatize_word(word) for word in x.split()]))
```
