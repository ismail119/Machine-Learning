```python
import pandas as pd,seaborn as sns,numpy as np
import warnings 
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
data = pd.read_csv('goodreads_data.csv') 
```

```python
data.head()
```

```python
texts = data[['Book','Genres']]
```

```python
import nltk
from nltk.corpus import stopwords
from textblob import Word
from textblob import TextBlob
```

```python
#Harflerin küçültülmesi
texts.Book = texts.Book.apply(lambda x: " ".join(x.lower() for x in x.split()))
texts.Genres= texts.Genres.apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
#noktalamalarin kaldirilmasi
texts.Book = texts.Book.str.replace("[^\w\s]","")
texts.Genres = texts.Genres.str.replace("[^\w\s]","")
```

```python
texts.Book = texts.Book.str.replace("\d","")
texts.Genres = texts.Genres.str.replace("\d","")
```

```python
stopwords = stopwords.words("english")
```

```python
texts.Book = texts.Book.apply(lambda x: " ".join(x for x in x.split() if x not in stopwords))
texts.Genres = texts.Genres.apply(lambda x: " ".join(x for x in x.split() if x not in stopwords))
```

```python
texts.Book = texts.Book.apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
texts.Genres = texts.Genres.apply(lambda x: " ".join(Word(word).lemmatize() for word in x.split()))
```

```python
texts
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
texts['combined_text'] = texts['Book'] + ' ' + texts['Genres']
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(texts['combined_text'])
```

```python
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
y = data.Avg_Rating
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(tfidf_matrix,y,test_size=0.3,random_state=42)
```

```python
#models

from sklearn.linear_model import LinearRegression,Lasso
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
models = dict(
    lr_model = LinearRegression(),
    lasso_model = Lasso(alpha=0.1),
    cb_model = CatBoostRegressor(),
    lgbm_model = LGBMRegressor(),
    rf_model = RandomForestRegressor()
)
```

```python
#fitting 
for model in models.values():
    model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error

for k,v in models.items():
    print("{} has RMSE => {}".format(k,np.sqrt(mean_squared_error(v.predict(x_test),y_test))))
```

```python
# LightGBM tuning
from sklearn.model_selection import GridSearchCV

lgbm_params = {"learning_rate":[0.5,0.1,0.01,1],"n_estimators":[20,40,100,200,500,1000],"max_depth":range(1,9)}
lgbm_grid = GridSearchCV(models["lgbm_model"],lgbm_params,cv=10,n_jobs=-1).fit(x_train,y_train)
```

```python
lgbm_tuned =LGBMRegressor(**lgbm_grid.best_params_).fit(x_train,y_train)
```

### RMSE Result

```python
np.sqrt(mean_squared_error(lgbm_tuned.predict(x_test),y_test))
```

```python
#comparing reals and predictions
compare_df = pd.DataFrame(list(y_test),lgbm_tuned.predict(x_test),columns=["prediction"])
```

```python
compare_df["real"] =  compare_df.index
```
